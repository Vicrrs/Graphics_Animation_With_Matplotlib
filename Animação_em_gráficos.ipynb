```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
# Funcao que fornece o lado direito dy/dt
def model(t, y, w0):
  x = y[0]
  v = y[1]
  dxdt = v
  dvdt = -w0*x #sqrt(k/m)
  return [dxdt, dvdt]
```

```python
x0 = 0.0 #posicao inicial
v0 = 1.0 #velocidade inicial
y0 = [x0, v0]
delta_t = (0, 10) #depois =10
```

```python
#y = odeint(model,y0,t,args=(1,)) # resolve a EDO
sol = solve_ivp(model,delta_t,y0,max_step=0.5,args=(1,)) # resolve a EDO
```

```python
print(sol) # detalhar a saida
```

```python
len(sol.t)
```

```python
print(sol.y[0]) # x(t)
print(sol.y[1]) # v(t)
```

```python
plt.rc('figure',figsize=(8,6))
plt.rc('lines',linewidth=4,linestyle='--',markersize=14)
plt.rc('axes',linewidth=1.5,labelsize=16)
plt.rc('xtick',direction='in',labelsize=18)
plt.rc('ytick',direction='in',labelsize=18)
plt.rc('legend',frameon='False',fontsize=18,loc='best')
```

```python
fig1,ax = plt.subplots()
ax.plot(sol.t,sol.y[0],marker='o',label='x(t)')
ax.plot(sol.t,sol.y[1],marker='^',label='v(t)')
ax.legend()
plt.show()
```

```python
fig2,ax2 = plt.subplots(1,2,figsize=(14,6))
ax2[0].plot(sol.t,sol.y[0],marker='o',markersize=12,color='black',label='numerico x(t)')
ax2[0].plot(sol.t,np.sin(sol.t),marker='^',markersize=10,linestyle='-',color='red',label='analitico x(t)')
ax2[0].legend()
ax2[1].plot(sol.t,sol.y[1],marker='^',markersize=12,color='black',label='v(t)')
ax2[1].plot(sol.t,np.cos(sol.t),marker='^',markersize=10,color='red',label='v(t)')
ax2[1].legend()
plt.show()
```
